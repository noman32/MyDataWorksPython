```python
url2 = "https://www.epicurious.com/search/pasta%20chilli"
response2 = requests.get(url2)
```

```python
response2.status_code
```

```python
page_content2 = response2.content.decode('utf-8')
```

```python
bs_data2 = BeautifulSoup(page_content2, 'html.parser')
```

```python
print(bs_data2.prettify())
```

```python
bs_data2.find("a")
```

```python
bs_data2.find_all("a")
```

```python
bs_data2.find_all("article")
```

```python
recipeList = bs_data2.find_all("article", class_ = "recipe-content-card")
```

```python
recipeList
```

```python
recipeList[0].getText()
```

```python
one_recipe = bs_data2.find("article", class_ = "recipe-content-card")
```

```python
one_recipe
```

```python
one_recipe.find("a").get("href")
```

```python
url_recipe = "https://www.epicurious.com" + one_recipe.find("a").get("href")
```

```python
url_recipe
```

```python
one_recipeContent = requests.get(url_recipe)
```

```python
one_recipeContentStr = one_recipeContent.content.decode('utf-8')
```

```python
bs_data_one_recipe = BeautifulSoup(one_recipeContentStr, 'html.parser')
```

```python
bs_data_one_recipe.find("div", class_ = "BaseWrap-sc-gjQpdd BaseText-ewhhUZ Description-cSrMCf iUEiRd ioVvSX fsKnGI")
```

```python
for i in range(0, 10):
    print("hello")
```

```python

for i in bs_data_one_recipe.find_all("div", class_ = "BaseWrap-sc-gjQpdd BaseText-ewhhUZ Description-cSrMCf iUEiRd ioVvSX fsKnGI"):
    print(i.getText())
  

```

```python
recipe = []

for i in bs_data_one_recipe.find_all("div", class_ = "BaseWrap-sc-gjQpdd BaseText-ewhhUZ Description-cSrMCf iUEiRd ioVvSX fsKnGI"):
    # print(i.getText())
    recipe.append(i.getText())

recipe
```

```python
bs_data_one_recipe.find("li", class_ = "InstructionListWrapper-dcpygI kUNDwd").getText()
```

- using requests lib -> get connected to the url and saved the content into a var (eg response)
- converted response to string (responce.content.decode)
- converted the output intp BS data structure
- used find and find_all to find any specific tag (with even class or id name)
- used get_text to get the text content
- find_all returns a list, so we ran a loop to get text from each element in the  list.
